### Train-Test-Validation Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_temp, X_test, Y_temp, Y_test = train_test_split(
    df.iloc[:, 0:125], df.iloc[:, 125], test_size=0.2, random_state=1
)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(
    X_temp, Y_temp, test_size=0.25, random_state=1
)
```

```python
train = pd.DataFrame(X_train.join(Y_train))
train.to_csv("../data/interim/train-split.csv")
```

```python
val = pd.DataFrame(X_val.join(Y_val))
val.to_csv("../data/interim/val-split.csv")
```

```python
test = pd.DataFrame(X_test.join(Y_test))
test.to_csv("../data/interim/test-split.csv")
```

### Data Cleaning - Converting input training data to numeric

```python
for column in train:
    train[column] = pd.to_numeric(train[column], errors="coerce")

train.dtypes
```

```python
total_NaN_count = 0
total_null_count = 0
for column in train:
    total_NaN_count += train[column].isna().sum()
    total_null_count += train[column].isnull().sum()

print("Total NaN count: {}".format(total_NaN_count))
print("Total Null count: {}".format(total_null_count))
```

### Data Cleaning - Filling Missing Values

```python
train.describe()
```

```python
train.fillna(method="ffill", inplace=True)
train.describe()
```

```python
total_NaN_count = 0
total_null_count = 0
for column in train:
    total_NaN_count += train[column].isna().sum()
    total_null_count += train[column].isnull().sum()

print("Total NaN count: {}".format(total_NaN_count))
print("Total Null count: {}".format(total_null_count))
```

### Feature Selection from training data

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
train.iloc[:, 0:125].corrwith(train.iloc[:,125])
```

```python
features = SelectKBest(score_func=chi2, k=50)
features.fit(train.iloc[:, 0:125], train.iloc[:, 125])
best_features = list(train.iloc[:, 0:125].columns[features.get_support()])

for column in train.iloc[:, 0:125]:
    if column not in best_features:
        train.drop(columns=column, inplace=True)
```

```python
best_features
```

```python
train.to_csv("../data/processed/train.csv")
```

# Model

```python
X_train = train.iloc[:, 0:50]
Y_train = train.iloc[:, 50]
```

```python
test = pd.read_csv("../data/interim/test-split.csv").iloc[:, 1:127]

for column in test.iloc[:, 0:125]:
    if column not in best_features:
        test.drop(columns=column, inplace=True)

test.shape
```

```python
for column in test:
    test[column] = pd.to_numeric(test[column], errors="coerce")
test.fillna(method="ffill", inplace=True)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gnb_model = GaussianNB().fit(X_train, Y_train)
```

```python
X_test = test.iloc[:, 0:50]
Y_test = test.iloc[:, 50]
Y_predict = gnb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test, Y_predict))
```

### SVM Model

```python
from sklearn.svm import SVC
svm_model = SVC(kernel="poly", degree=8).fit(X_train, Y_train)
```

```python
X_test = test.iloc[:, 0:50]
Y_test = test.iloc[:, 50]
Y_predict = svm_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test, Y_predict))
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5).fit(X_train, Y_train)
```

```python
X_test = test.iloc[:, 0:50]
Y_test = test.iloc[:, 50]
Y_predict = knn_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test, Y_predict))
```
